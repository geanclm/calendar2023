```python
# CALENDRIO 2023 de uma pgina no Pandas
# adaptado de @leo.akio
# by @geanclm in 31/12/2022
# update in 02/01/2023

# importar lib
import pandas as pd

df = pd.DataFrame(
    {
        'C': ['', '', '', '1', '2', '3', '4', '5', '6', '7'],
        'A': ['', '', '', '8', '9', '10', '11', '12', '13', '14'],
        'L': ['', '', '', '15', '16', '17', '18', '19', '20', '21'],
        'E': ['', '', '', '22', '23', '24', '25', '26', '27', '28'],
        'N': ['', '', '', '29', '30', '31', '', '', '', ''],        
        'D': ['MAI', '', '','Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sb', 'Dom'],
        'A.': ['AGO', '', ',', 'Ter', 'Qua', 'Qui', 'Sex', 'Sb', 'Dom','Seg'],
        'R': ['FEV', 'MAR', 'NOV','Qua', 'Qui', 'Sex', 'Sb', 'Dom','Seg','Ter'],
        '2.': ['JUN', '', '', 'Qui', 'Sex', 'Sb', 'Dom','Seg','Ter','Qua'],
        '0': ['SET', 'DEZ', '', 'Sex', 'Sb', 'Dom','Seg','Ter','Qua','Qui'],
        '2': ['ABR', 'JUL', '', 'Sb', 'Dom','Seg','Ter','Qua','Qui','Sex'],
        '3': ['JAN', 'OUT', '', 'Dom','Seg','Ter','Qua','Qui','Sex','Sb']
    },
    index=['(', ')', '[',']','{','}','|','l','!','-'])

def color(x):
    if x == 'Dom' or x == 'Sb':
        color = 'red'
    elif x == 'MAI' or x == 'AGO' or x == 'MAR' or x == 'DEZ' or x == 'JUL' or x == 'JAN' or x == 'OUT' or x == '31':
        color = 'blue'
    elif x == 'NOV' or x == 'JUN' or x == 'SET' or x == 'ABR' or x == '30':
        color = 'green'
    elif x == 'FEV' or x == '28':
        color = 'orange'        
    else:
        color = 'black'
    return 'color: %s' % color

df.style.highlight_between(color = 'lightblue', axis=0, subset = ['C','A','L','E','N']).highlight_between(color = 'lightyellow', axis=0, subset = ['D','A.','R','2.','0','2','3']).applymap(color)
```
